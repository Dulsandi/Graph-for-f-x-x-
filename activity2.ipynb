Ploting the graph for f(x)=x^2

```python
import numpy as np
import matplotlib.pyplot as plt

#define the finction
def f(x):
  return x**2

#create values for x
x=np.linspace(-10,10,100)
#calculate y values
y= f(x)

#plot
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("f(x)=x^2")
plt.title("Simple plot for f(x)=x^2")
plt.grid(True)
plt.show()
```
